TAHAP PREPROCESSING DATA

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import unicodedata
```

```python
dataset = pd.read_csv('SocialMediaTop100.csv', encoding='latin-1')

x = dataset[['FOLLOWERS', 'POSTS']].values
y = dataset['Platform'].values

print(x)
```

```python
# Fungsi untuk menghilangkan karakter non-ASCII
def remove_non_ascii(text):
    return ''.join(char for char in unicodedata.normalize('NFKD', text) if ord(char) < 128)

# Menerapkan fungsi remove_non_ascii pada kolom 'Platform'
y = dataset['Platform'].apply(remove_non_ascii).values

# Menampilkan y
print(y)
```

```python
from sklearn.impute import SimpleImputer
# Membuat dua buah SimpleImputer, masing-masing untuk kolom "FOLLOWERS" dan "POSTS"
imputer_followers = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer_posts = SimpleImputer(missing_values=np.nan, strategy='mean')

# Menerapkan imputasi pada kolom "FOLLOWERS" dan "POSTS"
imputer_followers.fit(x[:, 0:1])
imputer_posts.fit(x[:, 1:2])

x[:, 0:1] = imputer_followers.transform(x[:, 0:1])
x[:, 1:2] = imputer_posts.transform(x[:, 1:2])
```

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Memilih kolom "FOLLOWERS" dan "POSTS" saja dari x_train dan x_test
x_train_selected = x_train[:, [0, 1]]
x_test_selected = x_test[:, [0, 1]]

# Melakukan scaling pada kolom "FOLLOWERS" dan "POSTS" dari x_train dan x_test
x_train_scaled = sc.fit_transform(x_train_selected)
x_test_scaled = sc.transform(x_test_selected)

# Mengganti nilai x_train dan x_test dengan data yang telah di-scale
x_train[:, [0, 1]] = x_train_scaled
x_test[:, [0, 1]] = x_test_scaled
```

```python
print(x_train)
```

PELATIHAN MODEL K-NN

```python
dataset.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min()-1, stop=x_set[:, 0].max()+1, step=0.01),
                     np.arange(start=x_set[:, 1].min()-1, stop=x_set[:, 1].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
            alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue', 'yellow', 'purple')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue', 'yellow', 'purple'))(i), label = j)
plt.title('Klasifikasi Data dengan K-NN (Data Training)')
plt.xlabel('FOLLOWERS')
plt.ylabel('POSTS')
plt.legend()
plt.show()
```
